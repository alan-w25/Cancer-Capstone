Lets read in the data again once more now that it's in the correct format. 

```python
table_df = pd.read_csv('hmp_abundance_table.csv')
table_df.set_index('SN', inplace=True)

table_df.head(n = 10)
```

```python
table_df.info()
```

The first 5 rows of the df seem to be metadata. We want to separate these from the actual samples. 

```python
metadata = table_df.iloc[:8, :]
metadata.columns = metadata.loc['SRS']
srs = metadata.loc['SRS']
metadata.drop('SRS', axis=0, inplace=True)
if metadata.index.name == "SN":
    metadata.index.name = None
export_path = 'abundance_metadata.csv'
metadata.to_csv(export_path)

metadata.head(n = 10)
```

We need to find how many unique STSites there are for hte final matrix

```python
organ_sites = pd.Series(metadata.loc['STSite'].unique())
organ_sites
```

```python
len(organ_sites)
```

```python
type(organ_sites)
```

There are 17 organ sites that were sampled from. 

```python
abundance_data = table_df.iloc[8:, :]
abundance_data.columns = srs
if abundance_data.index.name == "SN":
    abundance_data.index.name = None
abundance_data = abundance_data.rename_axis(None, axis=1)
abundance_data = abundance_data.astype('float')
export_path = 'abundance_data.csv'
abundance_data.to_csv(export_path)

abundance_data.head(n = 10)
```

Create a new data frame where the rows will be the taxa. Take the average of any taxon from different genders same name. 

```python
abundance_data.dtypes
```

```python
abundance_data.info()
```

```python
transpose_abundance = abundance_data.transpose() 
transpose_abundance.head()
```

```python
transpose_abundance['OrganSite'] = metadata.loc['STSite']

numeric_columns = transpose_abundance.drop(columns='OrganSite').select_dtypes(include=[float, int]).columns
transposed_df = transpose_abundance[['OrganSite'] + list(numeric_columns)]

group_by_site = transposed_df.groupby('OrganSite').mean()
```

```python
organ_site_abundance = group_by_site.transpose()
organ_site_abundance = organ_site_abundance.rename_axis("taxa_name", axis=1)
organ_site_abundance.columns = map(str.lower, organ_site_abundance.columns)
organ_site_abundance.head()
```

```python
organ_site_abundance.info()
```

```python
output = 'final_organ_site_abundance.csv'
organ_site_abundance.to_csv(output)
```
